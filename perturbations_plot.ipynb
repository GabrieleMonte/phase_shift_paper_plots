```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os
from tqdm import trange
import matplotlib.pylab as pl
import matplotlib as mpl
from matplotlib.gridspec import GridSpec
from scipy.interpolate import interp1d, CubicSpline
from scipy.ndimage import gaussian_filter

plt.style.use('/home/gabriele/Desktop/PyCharm_env/mine.mplstyle')
plt.rcParams["axes.axisbelow"] = False
```

```python
##############################
###   LOAD RELEVANT DATA:  ###
##############################

#File Directory:
files_dir="/home/gabriele/Desktop/JupiterNotebook_projects/phase_shift_paper_plots/files/"

#Main Plots Directory:
main_plots_dir="/home/gabriele/Desktop/JupiterNotebook_projects/phase_shift_paper_plots/main_plots/"

phase_shift_Pi_dT0_at_all_zs=np.load(files_dir+"phase_shift_full_data_Pi_dT0.npy",allow_pickle=True)
dphis_dT0_mean_at_zs, dphis_dT0_std_at_zs, dT0_pks_fiducial_at_zs=phase_shift_Pi_dT0_at_all_zs[:,0,:].transpose()
dphis_Pi_mean_at_zs, dphis_Pi_std_at_zs, Pi_pks_fiducial_at_zs=phase_shift_Pi_dT0_at_all_zs[:,1,:].transpose()


phase_shift_Pi_dT0_at_rec=np.load(files_dir+"dphis_over_pi_dT0_Pi_at_rec.npy",allow_pickle=True)
dT0_pks_fiducial_at_rec, dphis_dT0_mean_at_rec, dphis_dT0_std_at_rec=phase_shift_Pi_dT0_at_rec[0,:].transpose()
Pi_pks_fiducial_at_rec, dphis_Pi_mean_at_rec, dphis_Pi_std_at_rec=phase_shift_Pi_dT0_at_rec[1,:].transpose()

Pi_dT0_vs_krs_at_rec=np.load(files_dir+"perturbations_dT0_Pi_vs_krs_at_rec.npy",allow_pickle=True)

phase_shift_PBT_at_z=np.loadtxt(files_dir+"phase_shift_perturbations_template_at_z.dat", skiprows=1)
PBT_zs, PBT_phi_infty_over_pi_at_zs, PBT_krs_star_over_pi_at_zs,PBT_xi_at_zs=phase_shift_PBT_at_z.transpose()


#############################################
### Perturbation Based Template Function: ###
#############################################
def f_k_rs_at_z(krs_over_pi,phi_infty_over_pi,krs_star_over_pi,xi):
    return phi_infty_over_pi/(1+(krs_over_pi/krs_star_over_pi)**xi)


#Global Variables:

Neff_less=np.linspace(1, 2.9, num=50)
Neff_more=np.linspace(3.1, 6, num=50)
N_eff=np.zeros(100)
N_eff[:50]=Neff_less
N_eff[50:]=Neff_more
```

```python
#############################################
############       FIGURE 5      ############
#############################################


fig1=plt.figure(figsize=(6,2.5),dpi=300)
gs1 = GridSpec(20,42,figure=fig1)
ax1 = fig1.add_subplot(gs1[1:10,1:18])
ax2 = fig1.add_subplot(gs1[10:19,1:18])
ax3 = fig1.add_subplot(gs1[1:-1,24:-1])
axz = fig1.add_subplot(gs1[1:-1,-1])
axNeff = fig1.add_subplot(gs1[1:-1,18])

colors = pl.cm.rainbow(np.linspace(0,1,len(N_eff)))

# colormap
cmap = plt.get_cmap('rainbow', len(N_eff))
# Normalizer
norm = mpl.colors.Normalize(vmin=np.min(N_eff), vmax=np.max(N_eff))

# creating ScalarMappable
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
ax1.tick_params(axis='both', which='major', labelsize=6.5)
ax1.tick_params(axis='both', which='minor', labelsize=6.5)
ax2.tick_params(axis='both', which='major', labelsize=6.5)
ax2.tick_params(axis='both', which='minor', labelsize=6.5)

linewidths=np.linspace(0.2,1.1,num=len(N_eff))[::-1]
for i in trange(len(N_eff)):
    krs=np.array(Pi_dT0_vs_krs_at_rec[i][0])
    dT0s_norm=np.array(Pi_dT0_vs_krs_at_rec[i][1])
    Pis_norm=np.array(Pi_dT0_vs_krs_at_rec[i][2])
    ax1.plot(krs/np.pi,dT0s_norm,ls='solid',color=colors[i],alpha=0.8,linewidth=linewidths[i])
    ax2.plot(krs/np.pi,Pis_norm,ls='solid',color=colors[i],alpha=0.8,linewidth=linewidths[i])

plt.colorbar(sm,cax=axNeff)

ax1.set_ylabel(r'$[\Theta_0+\Psi](k,z_{\rm{rec}})$',fontdict={'fontsize':7.5,'family':'serif'})
ax2.set_ylabel(r'$\Pi(k,z_{\rm{rec}})$',fontdict={'fontsize':7.5,'family':'serif'})
ax2.set_xlabel(r'$k\cdot r_s^{\rm{rec}}/\pi$',fontdict={'fontsize':8,'family':'serif'})
ax1.set_xscale('log')
ax2.set_xscale('log')
ax1.set_xticks([1,10])
ax2.set_xticks([1,10])
ax1.set_xticklabels([])
ax2.set_xticklabels(['1','10'])
ax1.tick_params(axis='both', which='both',direction='in',  labelsize=6.5,top=True, right=True,length=2, width=.5)
ax2.tick_params(axis='both', which='both',direction='in',  labelsize=6.5,top=True, right=True,length=2, width=.5)

axNeff.set_xticklabels([])
axNeff.set_xlabel(r'$N_{\nu}$',fontdict={'fontsize':8,'family':'serif'})

######################################################
###   Compute splines of PBT parameters, namely:   ###
###    (1) \phi__infty(z)/\pi                      ###
###    (2) [(k r_s)_\star](z)/\pi                  ###
###    (3) \xi(z)                                  ### 
######################################################

phi_infty_over_pi_of_z=CubicSpline(PBT_zs, PBT_phi_infty_over_pi_at_zs)
krs_star_over_pi_of_z=CubicSpline(PBT_zs, PBT_krs_star_over_pi_at_zs)
xi_of_z=CubicSpline(PBT_zs, PBT_xi_at_zs)

krs_over_pi_global_array=np.linspace(0,14,1000)
z_global_array=np.linspace(500,2100,500)
PBT_at_zs=np.array([f_k_rs_at_z(krs_over_pi_global_array,phi_infty_over_pi_of_z(zval),krs_star_over_pi_of_z(zval),xi_of_z(zval)) for zval in z_global_array])


#make colors of length of indices_to_plot as a gradient of blues
colors_pbt = pl.cm.autumn(np.linspace(0,1,len(z_global_array)))[::-1]
cmap = plt.get_cmap('autumn_r', len(z_global_array))
norm = mpl.colors.Normalize(vmin=np.min(z_global_array), vmax=np.max(z_global_array))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

phase_shift_full_data_PBT=[]
for j in range(len(z_global_array)):
    ax3.plot(krs_over_pi_global_array,PBT_at_zs[j],color=colors_pbt[j],linewidth=1,zorder=1)

#Phase shift data for dT0 and Pi at recombination:
zrec=1085
ax3.errorbar(dT0_pks_fiducial_at_rec,dphis_dT0_mean_at_rec,yerr=dphis_dT0_std_at_rec,color='k',marker='s',markersize=.5,ls='none',elinewidth=.5,alpha=0.9,zorder=1.5,label=r'$[\Theta_0+\Psi](z_{\rm{rec}})$')
ax3.errorbar(Pi_pks_fiducial_at_rec,dphis_Pi_mean_at_rec,yerr=dphis_Pi_std_at_rec,color='k',marker='^',markersize=.5,ls='none',elinewidth=.5,alpha=0.9,zorder=1.5,label=r'$\Pi(z_{\rm{rec}})$')
ax3.plot(krs_over_pi_global_array,f_k_rs_at_z(krs_over_pi_global_array,phi_infty_over_pi_of_z(zrec),krs_star_over_pi_of_z(zrec),xi_of_z(zrec)),linewidth=.6,zorder=1.3,c='k')
#ax3.plot(krs_over_pi_global_array,f_k_rs_at_z(krs_over_pi_global_array,0.06048352,  1.48229336, -1.93326931),linewidth=.6,zorder=1.3,c='k')

plt.colorbar(sm,cax=axz)
#ax3.legend(prop={'family':'serif','size':5},loc='lower right')
ax3.tick_params(axis='both', which='major', labelsize=6.5)
ax3.tick_params(axis='both', which='minor', labelsize=6.5)
axz.tick_params(axis='both', which='major', labelsize=6.5)
axz.tick_params(axis='both', which='minor', labelsize=6.5)
axz.set_xticklabels([])
ax3.legend(prop={'family':'serif','size':6},loc='upper left')
#ax1.set_ylim(0,0.076)
ax3.set_xlim(0,14)
ax3.set_xticks([1,3,5,7,9,11,13])

ax3.set_ylabel(r'$\delta\phi_{\nu}/\pi$ ',fontdict={'fontsize':8,'family':'serif'})
ax3.set_xlabel(r'$k r_s(z)/\pi$',fontdict={'fontsize':8,'family':'serif'})
axz.set_xlabel(r'$z$',fontdict={'fontsize':8,'family':'serif'})
ax3.text(4.5,0.01,r'$f_\phi{(k r_s)}=\frac{\phi_{\infty}}{1+\left[\frac{k r_s}{(k r_s)_\star}\right]^{\xi}}$',fontdict={'fontsize':8,'family':'serif'})

plt.savefig('main_plots/phase_shift_dT_Pi.png',bbox_inches='tight')
plt.savefig('main_plots/phase_shift_dT_Pi.pdf',bbox_inches='tight')
plt.show()
plt.close()
```
